```python
import pandas as pd
df=pd.read_csv('facebookdataset.csv')
print(df.head())
```

```python
print(df.tail())
```

```python
#subset1
subset1=df[['age','tenure','www_likes']]
subset1.head()
```

```python
#subset2
subset2=df[['gender','tenure','dob_year']]
subset2.head()
```

```python
#merge
subset3=pd.merge(subset1,subset2)
subset3.head()
```

```python
#sort
subset3.sort_values("gender",axis=0,ascending=True,inplace=True)
subset3
```

```python
#transpose
result = subset3.transpose()
print(result)
```
